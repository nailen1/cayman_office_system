```python
def get_ticker_bbg_of_ticker(ticker):
    return f"{ticker} KS Equity"


class OrderSheet:
    def __init__(self, date=None):
        self.date = date
        self.order = self.get_order()
        self.tickers = self.get_tickers()
        self.tickers_bbg = self.get_tickers_bbg()

    def get_raw_order(self):
        if self.date == None:
            df = open_df_order_latest()
        else:
            df = open_df_order_by_date(self.date)
        self.raw = df
        return df
    
    def get_order(self):
        df = self.get_raw_order()
        df = preprocess_df_order(df)
        self.order = df
        return df
    
    def get_tickers(self):
        df = self.order
        tickers = list(df['ticker'])
        self.tickers = tickers
        return tickers

    def get_tickers_bbg(self):
        tickers = self.get_tickers()
        tickers_bbg = [get_ticker_bbg_of_ticker(ticker) for ticker in tickers]
        self.tickers_bbg = tickers_bbg
        return tickers_bbg 
    
    def get_name_and_price_of_tickers(self):
        df_name_price = get_df_name_price_of_tickers_bbg(sheet.tickers_bbg)
        self.prices = df_name_price
        return df_name_price
    
    def get_priced_order(self):
        if not hasattr(self, 'prices'):
            self.get_name_and_price_of_tickers()
        prices = self.prices
        df = self.order.merge(prices, how='left', left_on='ticker', right_index=True)
        self.priced_order = df
        return df


class Orders:
    def __init__(self, file_folder=DIR_ORDER):
        self.file_folder = file_folder
        self.dates = self.get_dates()
        self.orders = self.get_orders()
        self.history = self.get_order_history()

    def get_dates(self):
        dates = get_dates_of_order(file_folder=self.file_folder)
        self.dates = dates
        return dates
    
    def get_orders(self):
        dct = {}
        for date in self.dates:
            OrderSheet(date=date)
            dct[date] = OrderSheet(date=date).order
        self.orders = dct
        return dct
    
    def get_order_history(self):
        self.get_orders()
        df = pd.concat(self.orders.values(), axis=0)
        self.history = df
        return df
    
    def get_holding(self):
        self.get_order_history()
        df = self.history.groupby('ticker').sum()
        self.holding = df
        return df
    
    def get_tickers(self):
        df = self.history
        tickers = list(df['ticker'])
        self.tickers = tickers
        return tickers

    def get_tickers_bbg(self):
        tickers = self.get_tickers()
        tickers_bbg = [get_ticker_bbg_of_ticker(ticker) for ticker in tickers]
        self.tickers_bbg = tickers_bbg
        return tickers_bbg 
    
    def get_name_and_price_of_tickers(self):
        df_name_price = get_df_name_price_of_tickers_bbg(sheet.tickers_bbg)
        self.prices = df_name_price
        return df_name_price
    
    def get_priced_history(self):
        if not hasattr(self, 'prices'):
            self.get_name_and_price_of_tickers()
        prices = self.prices
        df = self.history.merge(prices, how='left', left_on='ticker', right_index=True)
        df = df.rename(columns={'NAME': 'name', 'PX_LAST': 'price_asof'})
        cols_ordered = ['ticker', 'name', 'num', 'price_asof']
        df = df[cols_ordered]
        self.priced_history = df
        return df
    
    def get_holdings(self):
        df = self.history
        df = df.groupby(by='ticker').sum()
        self.holdings = df
        if not hasattr(self, 'prices'):
            self.get_name_and_price_of_tickers()
        prices = self.prices
        df = self.holdings.merge(prices, how='left', left_on='ticker', right_index=True)
        df = df.rename(columns={'NAME': 'name', 'PX_LAST': 'price_asof'})
        cols_ordered = ['name', 'num', 'price_asof']
        df = df[cols_ordered]
        df['evaluation'] = df['num'] * df['price_asof']
        self.holdings = df
        return df    
```

```python
import re

# def get_df_status(date):
#     date = date.replace('-', '')
#     yyyymmdd = date
#     try:
#         status = open_excel(file_name=f'Post-Instruction Portfolio Weight {date}.xlsx')
#     except:
#         try:
#             mmdd = yyyymmdd[4:]
#             status = open_excel(file_name=f'Post-Instruction Portfolio Weight {mmdd}.xlsx')
#         except Exception as e:
#             print(e)
#             status = None
#     return status

def scan_files_including_date_in_file_name_prefix(prefix, date, file_folder):
    date = date.replace('-', '')
    yyyymmdd = date
    regex = f'{prefix}.*{yyyymmdd}'
    file_names = scan_files_including_regex(file_folder=file_folder, regex=regex)
    if len(file_names) == 0:
        mmdd = yyyymmdd[4:]
        regex = f'{prefix}.*{mmdd}'
        file_names = scan_files_including_regex(file_folder=file_folder, regex=regex)
    if len(file_names) == 0:
        file_names = None
    if file_names != None:
        file_names = sorted(file_names)
    return file_names


def open_df_status_of_date(date, file_folder='dataset-status'):
    prefix = 'Post-Instruction'
    file_name = scan_files_including_date_in_file_name_prefix(prefix=prefix, date=date, file_folder=file_folder)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    print(f'- open: {file_name} in {file_folder}')
    return df

def open_df_status_latest(file_folder='dataset-status'):
    prefix = 'Post-Instruction'
    file_name = scan_files_including_regex(file_folder='dataset-status', regex=prefix)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    print(f'- open: {file_name} in {file_folder}')
    return df

def open_df_sellbuy_latest(file_folder='dataset-sellbuy'):
    prefix = 'LKEF Trade'
    file_name = scan_files_including_regex(file_folder='dataset-sellbuy', regex=prefix)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    print(f'- open: {file_name} in {file_folder}')
    return df

def open_df_sellbuy_of_date(date, file_folder='dataset-sellbuy'):
    prefix = 'LKEF Trade'
    file_name = scan_files_including_date_in_file_name_prefix(prefix=prefix, date=date, file_folder=file_folder)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    print(f'- open: {file_name} in {file_folder}')
    return df

def preprocess_df_status(status):
    df = status.iloc[:2, :3]
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    return df

def extract_text_in_parentheses(text):
    match = re.search(r'\((.*?)\)', text)
    if match:
        return match.group(1)
    return None

def get_date_ref_from_status_col(col_fx):
    m_raw, d_raw = extract_text_in_parentheses(col_fx).split(' ')[-1].split('/')
    mm = m_raw.zfill(2)
    dd = d_raw.zfill(2)
    yyyy = get_today()[:4]
    date = f'{yyyy}-{mm}-{dd}'
    return date

def get_num_from_usd_str(usd_str):
    try:
        usd = float(usd_str.replace('$', '').replace(',', ''))
    except:
        usd = None
    return usd

def get_df_info(df):
    cols_ref = list(df.columns)
    date = get_date_ref_from_status_col(cols_ref[-1])
    df.columns = ['fund_name', 'amount_raw', 'usdkrw']
    df['amount_usd'] = df['amount_raw'].apply(get_num_from_usd_str)
    df['amount_krw'] = df['amount_usd'] * df['usdkrw']
    df['date'] = date
    df = df.set_index('date')
    return df

def get_data_info(df_info):
    data = df_info.reset_index().to_dict(orient='records')
    return data

def get_usdkrw(df_info):
    usdkrw = get_data_info(df_info)[0]['usdkrw']
    return usdkrw

def get_amount_usd(df_info):
    amount_usd = get_data_info(df_info)[0]['amount_usd']
    return amount_usd

def get_amount_krw(df_info):
    amount_krw = get_data_info(df_info)[0]['amount_krw']
    return amount_krw

def get_ticker_bbg_of_ticker(ticker):
    ticker_bbg = f'{ticker} KS Equity'.replace(' KS KS', ' KS')
    return ticker_bbg

def get_ticker_of_ticker_bbg(ticker_bbg):
    ticker = ticker_bbg.replace(' Equity', '')
    return ticker

def get_current_price_of_ticker_bbg(bcon, ticker_bbg):
    fld = 'PX_LAST'
    df = bcon.ref(tickers=ticker_bbg, flds=[fld])
    df.index = df['ticker'].apply(get_ticker_of_ticker_bbg)
    df = df[['value']]
    df = df.rename(columns={'value': fld})
    return df

def get_current_prices_of_tickers_bbg(tickers_bbg):
    dfs = [get_current_price_of_ticker_bbg(ticker_bbg) for ticker_bbg in tickers_bbg]
    df = pd.concat(dfs, axis=0)
    return df

def get_df_account(status):
    df = status.iloc[3:, :]
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    df = df.dropna()
    df['ticker'] = df['Symbol']
    df = df.set_index('ticker')
    df.columns.name = ''
    return df


COLUMNS_asof = ['Name', 'Symbol', '# of Current Shares', 'Purchase Price',
       'Purchase Amount']

COLUMNS_NEW = ['Name', 'New Order']

COLUMNS_AFTER = ['Name', '# of After Order Shares', 'Last Price',
       'After Order Amount(KRW)', 'After Order Amount(USD)',
       'Portfolio Weight']

MAPPING_COLUMNS_asof = {
    'Name': 'name',
    'Symbol': 'ticker',
    '# of Current Shares': 'num_asof',
    'Purchase Price': 'price_purchase',
    'Purchase Amount': 'amount_asof'
}

MAPPING_COLUMNS_NEW = {
    'Name': 'name_order',
    'New Order': 'num_order'
}

MAPPING_COLUMNS_AFTER = {
    'Name': 'name',
    '# of After Order Shares': 'num_after',
    'Last Price': 'price_last',
    'After Order Amount(KRW)': 'amount_after_krw',
    'After Order Amount(USD)': 'amount_after_usd',
    'Portfolio Weight': 'weight'
}

def get_df_account_asof(df_account):
    df = df_account[COLUMNS_asof]
    df.columns = [MAPPING_COLUMNS_asof[col] for col in df.columns]
    return df

def get_df_account_order(df_account):
    df = df_account[COLUMNS_NEW]
    df.columns = [MAPPING_COLUMNS_NEW[col] for col in df.columns]
    return df

def sum_df_asof_and_order(df_asof, df_order):
    df = pd.concat([df_asof, df_order], axis=1)
    return df

def get_df_tobe(df_sum, df_info):
    df = df_sum.copy()
    df_tobe = df.copy()[['name']]
    df_tobe['num_tobe'] = df['num_asof'] + df['num_order']
    tickers_bbg = df.index.map(get_ticker_bbg_of_ticker)
    df_tobe['price_last'] = tickers_bbg.map(get_current_prices_of_tickers_bbg)
    df_tobe['amount_tobe_krw'] = df_tobe['num_tobe'] * df_tobe['price_last']
    df_tobe['amount_tobe_usd'] = df_tobe['amount_tobe_krw'] / get_usdkrw(df_info)
    df_tobe['weight_tobe'] = df_tobe['amount_tobe_krw'] / get_amount_krw(df_info)
    return df_tobe

def get_df_tobe(df_sum, df_info, df_after):
    df = df_sum.copy()
    df_tobe = df.copy()[['name']]
    df_tobe['num_tobe'] = df['num_asof'] + df['num_order']
    tickers_bbg = df.index.map(get_ticker_bbg_of_ticker)
    # df_tobe['price_last'] = tickers_bbg.map(get_asof_prices_of_tickers_bbg)
    df_tobe['price_last'] = df_after['price_last']
    df_tobe['amount_tobe_krw'] = df_tobe['num_tobe'] * df_tobe['price_last']
    df_tobe['amount_tobe_usd'] = df_tobe['amount_tobe_krw'] / get_usdkrw(df_info)
    df_tobe['weight_tobe'] = df_tobe['amount_tobe_krw'] / get_amount_krw(df_info)
    return df_tobe

def get_df_account_after(df_account):
    df = df_account[COLUMNS_AFTER]
    df.columns = [MAPPING_COLUMNS_AFTER[col] for col in df.columns]
    return df
```

```python
DIR_ORDER = './dataset-order'

MAPPING_COLUMNS_ORDER = {
    'Name': 'name',
    'Symbol': 'ticker',
    'Amount': 'num'
}

def open_excel(file_name, file_folder='data'):
    file_path = os.path.join(file_folder, file_name)
    return pd.read_excel(file_path)

def scan_files_order(file_folder=DIR_ORDER):
    prefix = 'Life Asset trade instruction'
    file_names = scan_files_including_regex(file_folder=file_folder, regex=prefix)
    return file_names

def extract_date_in_file_name(file_name):
    match = re.search(r'\b\d{8}\b', file_name)
    if match:
        return match.group(0)
    return None

def get_order_date_in_file_name(file_name, form='%Y-%m-%d'):
    date = extract_date_in_file_name(file_name)
    if form == '%Y-%m-%d':
        date = f'{date[:4]}-{date[4:6]}-{date[6:]}' if '-' not in date else date
    return date

def get_dates_of_order(file_folder=DIR_ORDER, form='%Y-%m-%d'):
    file_names = scan_files_order(file_folder=file_folder)
    dates = [get_order_date_in_file_name(file_name=file_name, form=form) for file_name in file_names]
    return dates

def open_df_order_latest(file_folder=DIR_ORDER):
    prefix = 'Life Asset trade instruction'
    file_name = scan_files_including_regex(file_folder=DIR_ORDER, regex=prefix)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    date = get_order_date_in_file_name(file_name)
    df['date'] = date
    df = df.set_index('date')
    print(f'- open: {file_name} in {file_folder}')
    return df

def open_df_order_by_date(date, file_folder=DIR_ORDER):
    prefix = 'Life Asset trade instruction'
    file_name = scan_files_including_date_in_file_name_prefix(prefix=prefix, date=date, file_folder=file_folder)[-1]
    df = open_excel(file_name, file_folder=file_folder)
    date = get_order_date_in_file_name(file_name)
    df['date'] = date
    df = df.set_index('date')
    print(f'- open: {file_name} in {file_folder}')
    return df

def preprocess_df_order(df_order):
    df = df_order.copy()
    cols_to_keep = ['Symbol', 'Amount']
    df = df[cols_to_keep]
    df.columns = [MAPPING_COLUMNS_ORDER[col] for col in df.columns]
    # df = df.set_index('ticker')
    return df

def get_df_order_latest(file_folder=DIR_ORDER):
    df_order = open_df_order_latest(file_folder=file_folder)
    df_order = preprocess_df_order(df_order)
    return df_order

def get_df_order_by_date(date, file_folder=DIR_ORDER):
    df_order = open_df_order_by_date(date, file_folder=file_folder)
    df_order = preprocess_df_order(df_order)
    return df_order

def merge_df_orders(dfs_order):
    df = pd.concat(dfs_order, axis=0)
    return df

def open_df_holding_latest(file_folder='dataset-holding'):
    prefix = 'dataset-cayman-holding'
    file_name = scan_files_including_regex(file_folder=file_folder, regex=prefix)[-1]
    date = pick_input_date_in_file_name(file_name)
    date = f'{date[:4]}-{date[4:6]}-{date[6:]}'
    df = open_df_in_file_folder_by_regex(file_folder=file_folder, regex=prefix)
    df.columns.name = date
    print(f'- open: {file_name} in {file_folder}')
    return df

def merge_order_to_holding(df_holding, df_order):
    df = df_holding.merge(df_order, how='outer', left_index=True, right_index=True, suffixes=('_asof', '_order'))
    df = df.fillna(0)
    df['num_tobe'] = df['num_asof'] + df['num_order']
    for col in df.columns:
        df[col] = df[col].astype(int)
    return df

def generate_df_holding_from_df_merge(df_merge):
    df = df_merge.copy()
    df = df[['num_tobe']]
    return df

class OrderSheet:
    def __init__(self, date=None):
        self.date = date
        self.order = self.get_order()
        self.tickers = self.get_tickers()
        self.tickers_bbg = self.get_tickers_bbg()

    def get_raw_order(self):
        if self.date == None:
            df = open_df_order_latest()
        else:
            df = open_df_order_by_date(self.date)
        self.raw = df
        return df
    
    def get_order(self):
        df = self.get_raw_order()
        df = preprocess_df_order(df)
        self.order = df
        return df
    
    def get_tickers(self):
        df = self.order
        tickers = list(df['ticker'])
        self.tickers = tickers
        return tickers

    def get_tickers_bbg(self):
        tickers = self.get_tickers()
        tickers_bbg = [get_ticker_bbg_of_ticker(ticker) for ticker in tickers]
        self.tickers_bbg = tickers_bbg
        return tickers_bbg 
    
class Orders:
    def __init__(self, file_folder=DIR_ORDER):
        self.file_folder = file_folder
        self.dates = self.get_dates()
        self.orders = self.get_orders()
        self.history = self.get_order_history()

    def get_dates(self):
        dates = get_dates_of_order(file_folder=self.file_folder)
        self.dates = dates
        return dates
    
    def get_orders(self):
        dct = {}
        for date in self.dates:
            OrderSheet(date=date)
            dct[date] = OrderSheet(date=date).order
        self.orders = dct
        return dct
    
    def get_order_history(self):
        self.get_orders()
        df = pd.concat(self.orders.values(), axis=0)
        self.history = df
        return df
    
    def get_holding(self):
        self.get_order_history()
        df = self.history.groupby('ticker').sum()
        self.holding = df
        return df
```
