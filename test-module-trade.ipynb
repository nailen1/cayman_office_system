```python
def get_df_timeseries_of_a_ticker_in_a_trade(trade_data):
    date = trade_data['date']
    ticker = trade_data['ticker']
    num_shares = trade_data['num_shares']
    df = get_df_timeseries_by_ticker(ticker=ticker)
    df = df[df.index >= date]
    df['num_shares'] = num_shares
    df[f'{ticker}: ({date}, {num_shares})'] = df['price_last'] * df['num_shares']
    return df
```

```python
from datetime import datetime, timedelta

def generate_date_list(start_date, end_date):
    # 입력 형식 감지
    if '-' in start_date:
        date_format = "%Y-%m-%d"
    else:
        date_format = "%Y%m%d"
    
    # 문자열을 datetime 객체로 변환
    start = datetime.strptime(start_date, date_format)
    end = datetime.strptime(end_date, date_format)
    
    # 날짜 목록 생성
    date_list = []
    current_date = start
    while current_date <= end:
        date_list.append(current_date.strftime(date_format))
        current_date += timedelta(days=1)
    
    return date_list
```

```python
import mysql.connector
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv()

user = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PASSWORD")
host = os.getenv("MYSQL_HOST")
database = os.getenv("MYSQL_DATABASE_NAME")

connection = mysql.connector.connect(user=user, password=password, host=host, database=database)
cursor = connection.cursor()

def get_price_of_date_by_ticker(ticker, date):
    connection = mysql.connector.connect(user=user, password=password, host=host, database=database)
    cursor = connection.cursor()
    ticker = ticker[:6]
    query = f"SELECT dt, pr1 FROM p_ks_{ticker} WHERE dt = '{date}'"
    cursor.execute(query)
    data = cursor.fetchall()
    price_of_date = data[-1][-1]
    return price_of_date
```
